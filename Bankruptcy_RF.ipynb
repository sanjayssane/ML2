```
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['Bankruptcy.csv']))
df.head()
```

**Making Necessary Imports**

```
import numpy as np
import warnings; warnings.simplefilter('ignore')
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

Instantiating the model

```
X = df.iloc[:,3:]
y = df.iloc[:,1]
model_rf = RandomForestClassifier(random_state=1211)
```

Listing the tuning parameter values

```
parameters = {'max_features': np.linspace(3,20,18).astype(int),
              'n_estimators': [100,200,300,400,500,600]}
print(parameters)
```

Setting the Grid Search

```
cv = GridSearchCV(model_rf, param_grid=parameters,cv=5,scoring='roc_auc')
```

Running the Grid Search

```
cv.fit( X , y )
```

Results:

```
results = pd.DataFrame(cv.cv_results_)
results
```

**Best** Parameter Combination:

```
print(cv.best_params_)
```

**Best** ROC AUC Score

```
print(cv.best_score_)
```
